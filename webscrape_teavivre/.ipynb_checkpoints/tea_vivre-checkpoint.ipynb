```python
from bs4 import BeautifulSoup
import urllib2

from email.MIMEText import MIMEText
import smtplib

GMAIL_LOGIN = 'onceuponatimeforever@gmail.com'
GMAIL_PASSWORD = 'Gilgamesh11223$'

def send_email(subject, message, from_addr=GMAIL_LOGIN, to_addr=GMAIL_LOGIN):
    msg = MIMEText(message)
    msg['Subject'] = subject
    msg['From'] = from_addr
    msg['To'] = to_addr
    msg['Reply-To'] = GMAIL_LOGIN

    server = smtplib.SMTP('smtp.gmail.com', 587)  # port 465 or 587
    server.ehlo()
    server.starttls()
    server.ehlo()
    server.login(GMAIL_LOGIN, GMAIL_PASSWORD)
    server.sendmail(from_addr, to_addr, msg.as_string())
    server.close()
    
def get_site_html(url):
    source = urllib2.urlopen(url).read()
    return source

def get_tree(url):
    source = get_site_html(url)
    tree = BeautifulSoup(source, "lxml")
    return tree

def extract_prices():
    # First, I'm going to identify the areas of the page I want to look at
    oolongs = get_tree('http://www.teavivre.com/oolong-tea/')
    o_dict = {}
    price_dict = {}
    all_teas_list = []

    for o_link in oolongs.find_all('h2', {'class':'product-name'}):
        if "Out of Stock" in o_link.text:
            break
        oolong_page = get_tree(o_link.a['href'])
        for index, package_option in enumerate(oolong_page.find_all(attrs={"class": "options-list"})):
            for package_type in package_option.children:
                # Place in tea name, url
                o_dict['full_tea_name'] = o_link.text
                o_dict['url'] = str(o_link.a['href'])
                # Place in packaging style, pricing info 
                if index == 0:
                    package_style = 'Zip Bag'
                else:
                    package_style = 'Tin'
                if package_type.find(attrs={'class':'productprice'}) is not None:
                    price = str(package_type.find(attrs={'class':'productprice'}).text.replace(u'\xa0', '').strip())
                    quantity = str(package_type.find('label').text.strip())
                else:
                    # This is a special sampler collection price
                    price = str(oolong_page.find(attrs={'class':'special-price'}).text.strip())
                    quantity = "5 sample sachets"
                if package_style not in price_dict:
                    price_dict = {package_style: {}}
                price_dict[package_style][quantity] = price
                o_dict.update(price_dict)
        all_teas_list.append(o_dict)
        o_dict = {}
        price_dict = {}
    return all_teas_list

def make_email(all_teas_list):
    # Now, let's make a mail message we can read:
    message = 'Hey Susan! \n\n\n'
    message += 'OMG, I found some tea for you.\n\n'
    
    for tea_dict in all_teas_list:
        message += '\n\nName of tea: %s \n' %(tea_dict['full_tea_name'])
        
        if 'Zip Bag' in tea_dict:
            message += "Zip Bag Options: \n"
            for quantity, price in tea_dict['Zip Bag'].iteritems():
                message += quantity + ': ' + price + "\n"

        if 'Tin' in tea_dict:
            message += "Tin Options: \n"
            for quantity, price in tea_dict['Tin'].iteritems():
                message += quantity + ': ' + price + "\n"

    message = message.encode('utf-8')
    # To read more about encoding:
    # http://diveintopython.org/xml_processing/unicode.html
    message = message.replace('\t', '').replace('\r', '')
    message += '\n\nXOXO,\n Your IPython Notebook Script'

    # And email it to ourselves!
    email = GMAIL_LOGIN
    send_email('Oolong Tea Prices from Teavivre ', message, from_addr=GMAIL_LOGIN,
               to_addr=email)

def main():
    all_teas_list = extract_prices()
    make_email(all_teas_list)
main()
```
